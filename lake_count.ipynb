```python
### 1번(위치정보 없는 데이터)
district_list = list([
    '강남구',
    '강동구',
    '강북구',
    '강서구',
    '관악구',
    '광진구',
    '구로구',
    '금천구',
    '노원구',
    '도봉구',
    '동대문구',
    '동작구', #
    '마포구',
    '서대문구',
    '서초구',
    '성동구',
    '성북구',
    '송파구',
    '양천구',
    '영등포구',
    '용산구',
    '은평구',
    '종로구', #
    '중구',
    '중랑구'])

for i, district in enumerate(district_list):
  print(i, district)
district_set = set(district_list)
```

```python
df = pd.read_csv("서울시 하천개황 정보.csv", encoding="cp949")
df_raw = pd.read_csv("서울시 하천개황 정보.csv", encoding="cp949", usecols=["하천명", "시점명", "종점명"])
df_raw
```

```python
#종점명에 지역구가 존재하면 해당 하천을 해당 지역구에 존재한다고 가정

for i, line in enumerate(df_raw["종점명"]):
  for word in line.split():
    if word in district_set:
      df_raw.loc[i,"종점지역구"] = word

df_raw = df_raw.dropna().drop(columns=["종점명"]).reset_index(drop=True)

print(df_raw)
```

```python
#시점명에 지역구가 존재하면 해당 하천을 해당 지역구에 존재한다고 가정

for i, line in enumerate(df_raw["시점명"]):
  for word in line.split():
    if word in district_set:
      df_raw.loc[i,"시점지역구"] = word

df_raw = df_raw.dropna().drop(columns=["시점명"]).reset_index(drop=True)

print(df_raw)
```
